Question 1. Produce a dispersion plot of the four main protagonists in Sense and Sensibility: Elinor, Marianne, Edward, and Willoughby. What can you observe about the different roles played by the males and females in this novel? Can you identify the couples?

```python
from nltk.corpus import gutenberg

sense_and_sensibility = nltk.Text(gutenberg.words('austen-sense.txt'))
```

```python
# Define the names of the protagonists
protagonists = ['Elinor', 'Marianne', 'Edward', 'Willoughby']

# Plot the dispersion of the protagonists
sense_and_sensibility.dispersion_plot(protagonists)
plt.title('Dispersion Plot of Protagonists in "Sense and Sensibility"')
plt.xlabel('Chapters')
plt.show()
```

Question 3. Use text9.index() to find the index of the word sunset. Youâ€™ll need to insert this word as an argument between the parentheses. By a process of trial and error, find the slice for the complete sentence that contains this word.

```python

text9 = gutenberg.words('austen-sense.txt')

try:
    sunset_index = text9.index('sunset')


    print("Index of 'sunset':", sunset_index)
    start_index = max(sunset_index - 10, 0)
    end_index = min(sunset_index + 10, len(text9))

    complete_sentence_slice = text9[start_index:end_index]

    print("Complete sentence containing 'sunset':", ' '.join(complete_sentence_slice))
except ValueError:
    print("The word 'sunset' is not present in the text.")
```

Question 4. What is the difference between the following two tests: w.isupper() and not w.islower()?

w.isupper() checks if all characters in w are uppercase.
not w.islower() checks if w contains at least one uppercase character or any non-alphabetic characters.
They both return True for strings that contain non-alphabetic characters or have no alphabetic characters.

w.isupper(): This method is used to check if all characters in the string w are uppercase. It returns True if all alphabetic characters in w are uppercase, and False otherwise. Here's a breakdown:
If all alphabetic characters in w are uppercase, w.isupper() returns True.
If there's at least one lowercase alphabetic character in w, w.isupper() returns False. It returns False for strings that contain non-alphabetic characters or have no alphabetic characters.

not w.islower(): This expression evaluates to True if the string w is not entirely lowercase. It checks if w contains any characters that are not lowercase. Here's a breakdown:
w.islower() returns True if all alphabetic characters in w are lowercase.
not w.islower() is the negation of this result, so it evaluates to True if w contains at least one uppercase character or any non-alphabetic characters.
It returns True if w contains uppercase characters or non-alphabetic characters, and False if all alphabetic characters in w are lowercase.

Question 5. Define a function percent(word, text) that calculates how often a given word occurs in a text and expresses the result as a percentage.

```python
def percent(word, text):
    word_count = text.count(word)
    total_words = len(text)

    if total_words > 0:
        word_percentage = (word_count / total_words) * 100
    else:
        word_percentage = 0

    return word_percentage

#using sample text for evaluation of the function
text = "This is a sample text with sample words. Sample text is used for demonstration."
word = "sample"

percentage = percent(word, text)
print(f"The word '{word}' occurs {percentage:.2f}% of the time in the text.")
```

The expression set(sent3) < set(text1) compares whether the set of unique elements in sent3 is a proper subset of the set of unique elements in text1. In other words, it checks if all elements in sent3 are also present in text1, and sent3 is not equal to text1.

Use cases for comparing sets using the < operator are:

Duplicate Removal: If you have a list of elements where duplicates are allowed, converting it to a set using set() will remove duplicates. You can then compare the original list with the set to check if any duplicates were present.

Data Validation: In data processing tasks, you might want to ensure that certain conditions are met before proceeding with further operations. Comparing sets can be a way to validate data integrity or check for expected patterns.

Feature Engineering: When working with machine learning models, you might need to create features based on the presence or absence of certain elements. Comparing sets can help in feature engineering by identifying relationships or similarities between different sets of data.
